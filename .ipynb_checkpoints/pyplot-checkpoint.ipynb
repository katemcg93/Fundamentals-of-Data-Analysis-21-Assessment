## What is Data Visualization?

Data visualization is the process of taking raw data and converting it to a format easily absorbed by the human brain, such as an image, chart or plot. The main goal of visualization is to help identify trends and patterns in data, as well as outliers. These insights can then be used to aid the decision-making process [1]. 

In the current era of Big Data, the ability to contextualize the trillions of rows of raw data generated daily is essential; visualisation is integral to this process. [2]. Data visualisation is useful for cleaning data and understanding its structure, detecting outliers or unexpected patterns, evaluating the validity of model output and presenting results. Images often reveal features of the dataset that are overlooked by statistics and models, such as unusual distributions of data, clusters, outliers and missing values [3]. 

As well as improving our understanding of a dataset, visualisation can help us communicate our findings more effectively. Good visualisation uses data to tell a story, filtering out the noise and keeping only the actionable insights. Easily understandable data benefits almost every industry: all areas of STEM, government, marketing, finance, sports, consumer goods and so on. Effective visualisation can help stakeholders from any field understand their data and use it to make informed decisions [2].

## Overview of Matplotlib

One of the most widely used Python graphing libraries is Matplotlib. Matplotlib allows users to create two-dimensional plots of data from a range of sources, including lists, numpy arrays and pandas dataframes. 

Matplotlib uses an object-oriented approach to plotting, meaning that plots can be built step by step by adding additional elements to them. This gives users a high level of control over the plots they create and lends itself to a high degree of customizabilty [5].

There are two main objects associated with a matplotlib plot:
1. Figure: The entire plot space, which can hold one or more axes
2. Axis : Axes are the individual plots that are rendered within the figure.

Note that in Matplotlib, the axes refer to the plot in its entirety, including the x and y axes, as illustrated below.

Almost all elements of a matplotlib plot, up to and including ticks and labels, has its own callable object and can therefore be customized.

## MATLAB and Pyplot

MATLAB is a high-performance language designed for mathematical computing. Some common use cases for MATLAB include:

 - Algorithm development
 - Modelling, simulating and prototyping
 - Data Analysis, exploration and visualization
 
MATLAB is well-established as a high-quality environment for computation and data analysis, particularly when working with arrays or matrices. However, in recent years Python has become increasingly popular for these purposes, and its computational power for mathematical tasks is on a par with MATLAB. Additionally, Python is open-source while MATLAB is proprietary and expensive, meaning a programme developed in MATLAB will reach a much smaller audience than one developed in Python. [5][6][7]

Hence, MATLAB programmers are converting to Python, and the pyplot module is designed to ease this transition by emulating MATLAB's global style. In MATLAB, the concept of importing is less prevalent than in Python, and most functions are available to the user at the top level, without the need for additional import commands. The main purpose of the pyplot module is to bring functions and classes from MATLAB and python directly into the namespace, which means that users can call them without additional import statements [6]. In older versions of matplotlib, a package named pylab was also available, which bulk imported matplotlib.plyplot and numpy for mathematical functions using the below command:

```python
from pylab import *
```

The advantage of using pylab is that the pyplot and numpy functionalities are combined in the same namespace/environment, making the envrionment even more similar to MATLAB. However, although pylab is still available in more recent versions, it is not recommended for use by matplotlib creators. The reason is that it can cause namespace pollution by overwriting many of Python's built in functions and causing them to behave in unexpected ways. Hence, it is recommended to use pyplot instead and import numpy separately. [5][6][7]

## Pyplot Object Hierarchy

A matplotlib plot consists of a hierarchy of Python objects [8][9]. 

- Figure: As already mentioned in this report, the Figure is the outermost object in the hierarchy, and is the container for all plot elements. It is the final image that contains one or more axes, and keeps track of all the axes. Note that the figure is only a container, and it is not possible to directly plot data onto it.
- Axes: Area on which data can be plotted. A single figure can contain multiple axes, but a given axes object can only be in one figure. The axes contain most of the other plot elements - axis, tick, plotlines, text etc.
- Axis : This object is similar to a number line, and takes care of setting graph limits, generating ticks (marks on the axis) and tick labels (strings that label the ticks). The location of the ticks is determined by a Locator object, and the tick labels are modified by a Formatter. When the Locator and Formatter objects are used in tandem, this gives a high degree of control over tick locations and labels. 
- Artist: The artist object encompasses everything that is visible on the rendered plot image, including the figure, Axes and Axes objects. When the plotted figure is rendered, all of the artists are drawn to the canvas. A given artist can only be in one axes object.

# State-based vs Object Oriented Plotting Approaches

Two methods of plotting are supported by matplotlib:
 - pyplot/state - based interface
 - Object-Oriented Interface (OO).
 
## State Based Interface
 
The state based interface is where the state (i.e. the figure) is preserved through various function calls, which are the methods that modify the figure. In other words, only one Figure or Axes are modified at a given time, without requiring them to be explicitly referenced [9][10]

This functionality is offered by the pyplot module. Almost all pyplot functions are either implicitly referencing an existing current figure/axes, or creating new ones if none exist. [11]

Below is an example of a simple plot generated using the state-based approach.

```python
import numpy as np

# Creating some data to plot
random_nos = np.arange (20)
plt.plot(random_nos)
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.legend(["Label"])
plt.title("A Basic Plot")
plt.show()
plt.close()
```

As seen above, we are able to generate a figure, plot some data and manipulate the plot elements without explicitly referring to the current figure/axes. 

The pyplot interface creates a new instance of these objects as none were available, and subsequent modifications to the figure are assumed to be part of the same instance. The plt.close() function at the end of the programme closes the figure window. [12]

Although the pyplot interface has some advantages, mainly concerning its ease of use, it is recommended that new adopters of matplotlib learn the object-oriented approach and stick to it. This is because the pyplot interface doesn't scale when creating multiple plots in the same figure, or for highly customized plots. The object-oriented interface gives users a greater degree of control over plot creation and customization. [13]

## Object - Oriented Interface

The object-oriented approach in matplotlib entails creating figure objects such as fig and ax and then calling methods (functions) on the objects themselves to manipulate specific elements of the plot. [4][14]

In this approach, Pyplot is used for certain functions, such as initializing the plot. The figure and axes objects are explicitly created and tracked by the user. The axes objects are used for most of the plotting actions.[15]

The below code gives an example of plotting using the object-oriented interface. The plot is initalized by calling the plt.subplots() function, and four separate axes objects are passed as arguments, along with the rows and columns the rendered figure should contain. A separate histogram is plotted and customized within each subplot by calling the individual ax instances.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random

#Setting seed so output will be constant
np.random.seed(1234)

#Generating data to plot
norm1 = np.random.normal(20,6,1000)
norm2 = np.random.normal(20,6,1000)
norm3 = np.random.normal(20,6,1000)
norm4 = np.random.normal(20,6,1000)

#Setting up fig and ax objects - 2 rows and two columns for 4 axes
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)

#This is one of the cases where pyplot is used in the object oriented approach - 
#calling a global title for all subplots rather than setting them individually
plt.suptitle("Hist Subplots")

#Plotting each of the four datasets on their own 
#ax and setting individual hist colours

ax1.hist(norm1, bins = 10, color = '#A0E8AF')
ax2.hist(norm2, bins = 10, color = '#86BAA1')
ax3.hist(norm3, bins = 10, color = '#EDEAD0')
ax4.hist(norm4, bins = 10, color = '#FFCF56')

#Hiding X and Y ticks and ticklabels on plots 2 and 3 to make the image less cluttered.
ax2.set_yticks([])
ax3.set_yticks([])

ax2.set_yticklabels([])
ax3.set_yticklabels([])

ax2.set_xticks([])
ax3.set_xticks([])

ax2.set_xticklabels([])
ax3.set_xticklabels([])

plt.show()
```

## 2.1 Stacked Histogram

A stacked histogram is where two or more histograms are displayed on the same scale, and is used to compare one categorical and one  quantitative variable. Unlike a single histogram, where only one quantitative variable can be analysed at a time, the stacked variable takes a categorical variable into account and separates quantitative data for each category into its own histogram, allowing for direct comparison between distinct groups. [19][20]

This histogram type could be used to determine the degree to which different populations differ regarding a given variable. The example below takes data from the Framingham Heart study, which is a longitudal, population-based cohort study initiated by the US Public Health Service to investigate risk factors for cardiovascular disease. Data gathered on the participants included gender, age, smoking status, BMI, highest level of education attained, blood pressure, glucose levels and whether the participant was at risk of developing CHD within the next 10 years (yes/no) [21].The dataset is currently available on Kaggle [22]. 

Participants were divided into four categories based on their education level, with some secondary education being the lowest and completing third-level education being the highest. The below code generates a stacked histogram to show the distribution of BMI amongst the participants, categorized by education level, by creating four separate dataframes and plotting the BMI data for each on the same figure, and setting the Stacked keyword argument to True, so that the histograms will be stacked on top of each other rather than side by side.

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv

#Changing style of plots to look more like Seaborn
plt.style.use('seaborn-muted')

#Only require two decimal places for output
pd.set_option("display.precision", 2)

#Reading the csv file containing framingham data
fh_dataframe = pd.read_csv('Archive/Framingham.csv')

#Separating Education levels into their own datafrmames, so they can be separately accessed/plotted
eduLevel1 = fh_dataframe[fh_dataframe["education"] == 1]
eduLevel2 = fh_dataframe[fh_dataframe["education"] == 2]
eduLevel3 = fh_dataframe[fh_dataframe["education"] == 3]
eduLevel4 = fh_dataframe[fh_dataframe["education"] == 4]

#Plotting each histogram on the same figure and setting stacked to true
#Labels willl be used to create the legend
fig, ax = plt.subplots()
ax.hist(eduLevel1["BMI"], stacked = True, label = 'Some Second Level',  bins = 15)
ax.hist(eduLevel2["BMI"], stacked = True, label = 'Completed Second Level',  bins = 15)
ax.hist(eduLevel3["BMI"], stacked = True, label = 'Some Third Level',  bins = 15)
ax.hist(eduLevel4["BMI"], stacked = True, label = 'Completed Third Level',  bins = 15)

#Hiding top and right spines so only x and y axis are visible
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Setting title and showing legend
plt.suptitle("Framingham Study - BMI Distribution by Education Level")
legend = plt.legend()

plt.show()
plt.close()
```

The above plot gives us several pieces of key information about the dataset. The first is that the largest cohort of participants are those with some second level education, followed by those who completed second-level, which is logical as this study was initiated in 1948 and mainly focused on middle-aged people [21].

The second is that based on this distribution, BMI appears to be inversely related to education level. The plots for those who had some or completed third level education are concentrated around the 20-30 BMI range, with a peak at around 22, and a relatively small proportion of these individuals appear to be in the obese category (BMI > 30). In contrast, those with some/completed secondary education have a higher proportion of individuals in the underweight (BMI <18.5), overweight (>25) and obese (>30) categories. This suggests that those who continued their education past second level are more likely to be at a helathy weight than those who did not. However, further analysis would be required before a true relationship between these two factors could be inferred. 

Thus, this histogram provides a clearer picture on the the education level of the participants and potentially their health status, and allows us to draw direct comparisons between the four cohorts. 

## 2.2 Correlation Matrix

A correlation matrix is a table that shows the relationship between given variables in a dataset. Each cell in the matrix represents the strength of the correlation between two variables [23].

The relationship between the pairs of variables is quantified by the Pearson Correlation Coefficient. This is a measure of the linear association between two variables, with a value of -1 to 1, where [24]:
- -1 indicates a perfect negative correlation between the variables.
- 0 indicates no correlation between the variables. 
- +1 indicates a perfect positive correlation.

The further away on either side the correlation coefficient is from 0, the stronger the relationship between the two variables in question. 

The main applications of the correlation matrix are as follows [25]: 
- Summarize large amounts of data where the goal is to quickly determine whether patterns may exist, i.e. whether the variables correlate with each other
- In conjunction with other statistical analyses, such as multiple linear regression, which is used to assess the relationship between two or more independent variables and one output variable. In such a scenario, the correlation matrix may be used to determine whether a relationship exists between the independent variables.

The below correlation matrix is created using matplotlib's colormap(cmap). Colormaps are simply a dictionary that maps numerical data to colours. The colors in the map are typically contrasting/divergent, and used to differentiate/distingush data in the plot . [26]

The main reason to use a colormap is that it is easier for humans to use colors to differentiate datapoints using colours than numerical values. [27]

In the case of this matrix, a strong positive correlation is represented by a red shade, and a strong negative correlation by a purple colour, and the strength of the color in an indvidual cell corresponds to the strength of the relationship, positive or negative, between the two variables. 

```python
#Fixing the gender column for corr matrix - it should say gender, not male
fh_dataframe = fh_dataframe.replace({'male':1}, value = 'Male')
fh_dataframe = fh_dataframe.replace({'male':0}, value = 'Female')
fh_dataframe = fh_dataframe.rename(columns= {'male':'gender'})

plt.style.use('default')

#Calculates correlation coefficient for all variables in dataset
corr_coeff_fh = fh_dataframe.corr()

#Making figure bigger so plot can be seen clearly
fig = plt.figure(figsize = (20,20))

#Calling suplot - only one required
ax = fig.add_subplot(111)

#Plotting correlation coefficient
corrmat = ax.matshow(corr_coeff_fh,cmap = 'rainbow', vmin = -1, vmax = 1)
ax.grid(False)
fig.colorbar(corrmat)
ticks = np.arange(0, len(corr_coeff_fh.columns), 1)
ax.set_xticks(ticks)
plt.title("Correlation Matrix - Framingham Heart Study", fontsize = 18)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(corr_coeff_fh.columns)
ax.set_yticklabels(corr_coeff_fh.columns)
plt.show()
plt.close()
```

The above plot shows the correlation between each pair of variables in the Framingham Heart Study. The colour scale starts at 1, which represents a perfect positive correlation, and ends at -1 for a perfect negative correlation. The correlation coefficents along this scale are mapped to a point on the color gradient (purple -> red).

The correlation matrix gives the following insights into the variables and how they relate to each other:
- Several of the variables appear to be negatively correlated with age, such as education level, smoking status and heart rate (resting heart rate tends to decrease with age). Additionally, blood pressure and cholesterol appear to increase with age.
- Education and smoking status both appear to be negatively correlated with BMI.
- As may be expected, strong correlations exist between the smoker/non-smoker and cigarettes per day variables, systolic and diastolic blood pressure and the presence of hypertension and glucose and the onset of diabetes.
- Although some variables are negatively correlated with ten-year CHD risk (education, heart rate) and some are positively correlated (age, hypertension, systolic blood pressure), the correlations are relatively weak, suggesting that CHD risk may be determined by a number of factors, rather than one factor in particular. 

As seen in the above example, the correlation matrix provides an effective method of visualizing the relationships that exist between variables in a complex dataset. 

## 2.3 Stacked Area Chart

A stacked area chart combines the line and bar chart, and shows how values relating to one or more numeric group change over the progression of a second variable, usually time [28].

The values of each group are displayed on top of each other, which allows us to see on the same figure the total evolution of a variable, and the contribution of each group to this evolution [29]

Stacked area charts can be created with the following matplotlib function [30]:
 - plt.stackplot(x,y)

Where:
- x: numeric variable used to measure progression, usually time
- y: variable/groups of interest

The below code creates a stacked area plot in matplotlib displaying the Covid-19 case numbers by continent over the last 4 weeks (weeks 46 - 49 2021). First, a csv file from the ECDC website containing the recorded case figures is read, and the last four weeks are isolated into a dataframe [31]. 

The file includes cases by country and continent, so next the program filters on overall continent figures - Africa, Asia, Europe, America and Oceania, and stores the weekly case numbers for each continent as a numpy array.

Finally, the stacked area plot is generated, with time (weeks) on the x-axis, and the covid case figures by continent on the y - axis.

```python
#Reading csv containing data from Eurostat on Weekly Covid Cases
covid_df = pd.read_csv("Archive/Covid Case Data Eurostat.csv")

#File has cases and deaths, isolating data on cases to plot
covid_df = covid_df.loc[covid_df["indicator"] == "cases"]

#Creating new column with numeric form of year - week to filter on last 4 weeks
covid_df["Year_Week_int"] = covid_df["year_week"].str.replace("-", "")
covid_df["Year_Week_int"] = pd.to_numeric(covid_df["Year_Week_int"])
covid_4_wks = covid_df.loc[covid_df["Year_Week_int"] > 202145]

#Isolating the total weekly case numbers for each continent into its own dataframe to plot later
covid_africa = covid_4_wks.loc[covid_4_wks["country"] == "Africa (total)"]
covid_europe = covid_4_wks.loc[covid_4_wks["country"] == "Europe (total)"]
covid_america = covid_4_wks.loc[covid_4_wks["country"] == "America (total)"]
covid_asia = covid_4_wks.loc[covid_4_wks["country"] == "Asia (total)"]
covid_oceania= covid_4_wks.loc[covid_4_wks["country"] == "Oceania (total)"]

#Retrieving weekly count figures for each continent
y1 = covid_africa["weekly_count"]
y2 = covid_europe ["weekly_count"]
y3 = covid_america ["weekly_count"]
y4 = covid_asia ["weekly_count"]
y5 = covid_oceania["weekly_count"]

#Setting up x axis - bins and tick labels
x = [0,1,2,3]
tick_labels = ["Wk 46", "Wk 47", "Wk 48", "Wk 49"]

#Labels for legent
labels = ["Africa", "Europe", "America", "Asia", "Oceania"]

#Converting the pandas columns with weekly case numbers to numpy arrays
array1 = y1.to_numpy()
array2 = y2.to_numpy()
array3 = y3.to_numpy()
array4 = y4.to_numpy()
array5 = y5.to_numpy()

plt.style.use('seaborn-pastel')

fig, ax = plt.subplots()
plt.suptitle("Weekly Covid Cases by Continent")
plt.title("Weeks 46 - 49 2021")
ax.stackplot(x, array1, array2, array3, array4, array5, labels = labels)
ax.locator_params(tight=True, nbins=4)
ax.set_xticks(range(len(tick_labels)))
ax.set_xticklabels(labels = tick_labels)

ax.set_xlabel("Week Number")
ax.set_ylabel("Weekly Covid Cases in Millions")

#Hiding top and right spines so only x and y axis are visible
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#Adding a tick label at each 1m interval
ax.set_yticks(np.arange(0, 5000000, 1000000))
plt.legend()
plt.show()
plt.close()
```

The above plot illustrates the overall progression of Covid cases over the last 4 weeks, as well as the relative contributions of each country. For example, we can see there was a spike in the overall number of cases between weeks 47 and 48, which likely relates to the spread of the Omnicron variant. Additionally, we can see that Europe accounts for more than half of the weekly covid cases, and that Africa, Asia and Oceania combined account for less than a quarter. However, in Africa the cases appear to have increased steadily over the last 4 weeks, which again may be due to the Omnicron variant, which was detected in South Africa.

In summary, matplotlib is a powerful tool for data visualization. As the above examples illustrate, the pyplot package allows us to create visually impactful plots in just a few lines of code, that provide us with valuable insights into a dataset.
